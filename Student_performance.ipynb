```python
import pandas as pd

df = pd.read_csv("student_performance_clean.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
df.describe()
```

```python
df.sort_values(by="Average", ascending=False).head(10)
```

```python
df[df["Result"] == "Fail"]
```

```python
df.groupby("Department")["Average"].mean()
```

```python
import matplotlib.pyplot as plt

df["Grade"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Grade Distribution")
plt.show()
```

```python
df.groupby("Department")["Average"].mean().plot(kind="bar")
plt.title("Department Wise Performance")
plt.show()
```

```python
df["Result"].value_counts().plot(kind="bar")
plt.title("Pass vs Fail Students")
plt.show()
```
