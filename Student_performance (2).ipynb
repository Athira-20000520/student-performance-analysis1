```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.columns = df.columns.str.strip()
```

```python
subject_avg = df[['Maths','Physics','Chemistry','English','Computer']].mean()
subject_avg
```

```python
df["Total"] = df[['Maths','Physics','Chemistry','English','Computer']].sum(axis=1)
df["Average"] = df["Total"] / 5
df.head()
```

```python
def grade(m):
    if m >= 90:
        return "A"
    elif m >= 75:
        return "B"
    elif m >= 50:
        return "C"
    else:
        return "D"

df["Grade"] = df["Average"].apply(grade)
df.head()
```

```python
df["Result"] = df["Average"].apply(lambda x: "Pass" if x >= 50 else "Fail")
df.head()
```

```python
topper = df.sort_values(by="Total", ascending=False).head(5)
topper
```

```python
weak = df.sort_values(by="Average").head(5)
weak

```

```python
subject_avg.plot(kind='bar', title="Subject-wise Average Marks")
plt.ylabel("Average Marks")
plt.show()
```

```python
df["Result"].value_counts().plot(kind="pie", autopct="%1.1f%%", title="Pass vs Fail")
plt.ylabel("")
plt.show()
```

```python
df[['Name','Total']].set_index("Name").head(10).plot(kind="bar", figsize=(10,5))
plt.title("Student Performance Comparison")
plt.ylabel("Total Marks")
plt.show()
```
